```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vaccine = pd.read_csv('features.csv')
vaccine.head()
```

```python
vaccine1 = pd.read_csv('labels.csv')
vaccine1.head()
```

```python
vaccine.info()
```

```python
vaccine1.info()
```

```python
#Data Integration
vaccine3 = vaccine.merge(vaccine1, how='left', on='respondent_id')
```

```python
vaccine3.head()
```

```python
#Missing Values
vaccine3.info()
```

```python
vaccine3.isna().sum()
```

```python
# Dropping Columns containing many Null values
vaccine3.drop(['health_insurance','employment_industry','employment_occupation'],axis=1,inplace = True)
```

```python
vaccine3.head()
```

```python
vaccine3.describe()
```

```python
vaccine3_cols = vaccine3.select_dtypes(include = 'object').columns
vaccine[vaccine3_cols].head()
```

```python
for col in vaccine3.columns:
    if vaccine3[col].isnull().sum() and vaccine3[col].dtypes != 'object':
        vaccine3[col].loc[(vaccine[col].isnull())] = vaccine3[col].median()
for col in vaccine3.columns:
    if vaccine3[col].isnull().sum() and vaccine3[col].dtypes == 'object':
        vaccine3[col].loc[(vaccine3[col].isnull())] = vaccine3[col].mode().max()
```

```python
vaccine3.isnull().sum()
```

```python
#Exploratory Data Analysis
# Target variable distribution
vaccine3['h1n1_vaccine'].value_counts(normalize=True).plot(kind='bar');
```

```python
# Gender and h1n1 vaccination status bar graph
sns.barplot(x=vaccine3['sex'], y=vaccine3['h1n1_vaccine']);
```

```python
# Education level percentage distribution of the vaccinated people 
sns.barplot(x=vaccine3['education'], y=vaccine3['h1n1_vaccine']*100);
```

```python
# Distribution of health workers who got the vaccine
sns.barplot(x=vaccine3['health_worker'], y=vaccine3['h1n1_vaccine']*100);
```

```python
# Doctor recommendation vs H1N1 vaccine
sns.barplot(x=vaccine3['doctor_recc_h1n1'], y=vaccine3['h1n1_vaccine']*100);
```

```python
# Setting the size of the plot
plt.figure(figsize = (5,5))
# Creating the bar plot with X = h1n1_knowledge and Y = h1n1_vaccine
sns.barplot(data = vaccine3, x=vaccine3['h1n1_knowledge'], y=vaccine3['h1n1_vaccine'])
# Setting title and labels
plt.title('H1N1 Knowledge vs Vaccination', fontsize=16)
plt.xlabel('H1N1 Knowledge', fontsize=13)
plt.ylabel('H1N1 Vaccination', fontsize=13)
plt.tight_layout();
```

```python
# Setting the size of the plot
plt.figure(figsize = (5,5))
# Creating the bar plot with X = h1n1_knowledge and Y = h1n1_concern
sns.barplot(data = vaccine3, x=vaccine3['h1n1_concern'], y=vaccine3['h1n1_vaccine'])
# Setting title and labels
plt.title('H1N1 Concern vs Vaccination', fontsize=16)
plt.xlabel('H1N1 Concern', fontsize=13)
plt.ylabel('H1N1 Vaccination', fontsize=13)
plt.tight_layout();
```

```python
plt.figure(figsize=(20,25))
plotnumber=1

for column in vaccine3:
    plt.subplot(3,3,plotnumber)
    sns.histplot(vaccine3[column])
    plt.xlabel(column,fontsize=20)
    plt.ylabel('Count',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
# Correlation matrix to see the relationship between numerical features
correlation = vaccine3.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="RdYlGn", fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
