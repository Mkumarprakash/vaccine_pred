```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vaccine = pd.read_csv(r'F:\features.csv')
vaccine.head()
```

```python
vaccine1 = pd.read_csv(r'F:\labels.csv')
vaccine1.head()
```

```python
vaccine.info()
```

```python
vaccine1.info()
```

```python
#Data Integration
vaccine3 = vaccine.merge(vaccine1, how='left', on='respondent_id')
```

```python
vaccine3.head()
```

```python
#Missing Values
vaccine3.info()
```

```python
vaccine3.isna().sum()
```

```python
# Dropping Columns containing many Null values
vaccine3.drop(['employment_industry','employment_occupation'],axis=1,inplace = True)
```

```python
vaccine3.head()
```

```python
vaccine3.describe()
```

```python
vaccine3_cols = vaccine3.select_dtypes(include = 'object').columns
vaccine[vaccine3_cols].head()
```

```python
for col in vaccine3.columns:
    if vaccine3[col].isnull().sum() and vaccine3[col].dtypes != 'object':
        vaccine3[col].loc[(vaccine[col].isnull())] = vaccine3[col].median()
for col in vaccine3.columns:
    if vaccine3[col].isnull().sum() and vaccine3[col].dtypes == 'object':
        vaccine3[col].loc[(vaccine3[col].isnull())] = vaccine3[col].mode().max()
```

```python
# Model Creation
X=vaccine3.drop(['h1n1_vaccine','seasonal_vaccine'], axis=1)
y_h1n1=vaccine3['h1n1_vaccine']
y_seasonal = vaccine3['seasonal_vaccine']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train_h1n1, y_test_h1n1, y_train_seasonal, y_test_seasonal = train_test_split(X,y_h1n1, y_seasonal, test_size=0.20, random_state=100) 
```

```python
#Logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression() #For H1N1
lr.fit(X_train,y_train_h1n1) 
```

```python
y_pred_h1n1 = lr.predict(X_test)
```

```python
lr.fit(X_train, y_train_seasonal)## training model for seasonal_vaccine
```

```python
y_pred_seasonal = lr.predict(X_test)
```

```python
#Evaluating Logistic Regression Model
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
# h1n1_vaccine

acc_h1n1 = accuracy_score(y_test_h1n1, y_pred_h1n1)
prec_h1n1 = precision_score(y_test_h1n1, y_pred_h1n1)
recall_h1n1 = recall_score(y_test_h1n1, y_pred_h1n1)
f1_h1n1 = f1_score(y_test_h1n1, y_pred_h1n1)
```

```python
# seasonal_vaccine

acc_seasonal = accuracy_score(y_test_seasonal, y_pred_seasonal)
prec_seasonal = precision_score(y_test_seasonal, y_pred_seasonal)
recall_seasonal = recall_score(y_test_seasonal, y_pred_seasonal)
f1_seasonal = f1_score(y_test_seasonal, y_pred_seasonal)
```

```python
print("For h1n1_vaccine:")
print(f"Accuracy: {acc_h1n1}")
print(f"Precision: {prec_h1n1}")
print(f"Recall: {recall_h1n1}")
print(f"F1 Score: {f1_h1n1}")
print("\n")
print("For seasonal_vaccine:")
print(f"Accuracy: {acc_seasonal}")
print(f"Precision: {prec_seasonal}")
print(f"Recall: {recall_seasonal}")
print(f"F1 Score: {f1_seasonal}")
print("\n")
```

```python
cr_h1n1=classification_report(y_test_h1n1, y_pred_h1n1)
print(cr_h1n1)
```

```python
cr_seasonal=classification_report(y_test_seasonal, y_pred_seasonal)
print(cr_seasonal)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
# Compute the ROC curve
fpr, tpr, thresholds = roc_curve(y_test_h1n1, y_pred_h1n1)

# Compute the AUC score
auc_score = roc_auc_score(y_test_h1n1, y_pred_h1n1)
print(f'AUC Score: {auc_score}')

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
# Compute the ROC curve
fpr, tpr, thresholds = roc_curve(y_test_seasonal, y_pred_seasonal)

# Compute the AUC score
auc_score = roc_auc_score(y_test_seasonal, y_pred_seasonal)
print(f'AUC Score: {auc_score}')

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
