```python
# Read the features.csv into a DataFrame
features_df = pd.read_csv('features.csv')

# Filling null values with mode, mean, or median for different columns

# For categorical columns, like 'education', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region', and 'census_msa', you can fill with mode
categorical_cols = ['education', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa', 'health_insurance']
for col in categorical_cols:
    features_df[col].fillna(features_df[col].mode()[0], inplace=True)

# For numerical columns, like 'h1n1_concern', 'h1n1_knowledge', 'behavioral_antiviral_meds', 'behavioral_avoidance', 'behavioral_face_mask', 'behavioral_wash_hands', 'behavioral_large_gatherings', 'behavioral_outside_home', 'behavioral_touch_face', 'doctor_recc_h1n1', 'doctor_recc_seasonal', 'chronic_med_condition', 'child_under_6_months', 'health_worker', 'opinion_h1n1_vacc_effective', 'opinion_h1n1_risk', 'opinion_h1n1_sick_from_vacc', 'opinion_seas_vacc_effective', 'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'household_adults', and 'household_children', you can fill with mean or median
numerical_cols = ['h1n1_concern', 'h1n1_knowledge', 'behavioral_antiviral_meds', 'behavioral_avoidance', 'behavioral_face_mask', 'behavioral_wash_hands', 'behavioral_large_gatherings', 'behavioral_outside_home', 'behavioral_touch_face', 'doctor_recc_h1n1', 'doctor_recc_seasonal', 'chronic_med_condition', 'child_under_6_months', 'health_worker', 'opinion_h1n1_vacc_effective', 'opinion_h1n1_risk', 'opinion_h1n1_sick_from_vacc', 'opinion_seas_vacc_effective', 'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'household_adults', 'household_children']
for col in numerical_cols:
    # Choose whether to fill with mean or median based on the distribution of data
    # If the data is skewed, use median; otherwise, use mean
    if features_df[col].skew() > 1:  # Example threshold for skewness
        features_df[col].fillna(features_df[col].median(), inplace=True)
    else:
        features_df[col].fillna(features_df[col].mean(), inplace=True)

# Remove specified columns
columns_to_remove = ['employment_occupation', 'employment_industry', 'employment_status', 'hhs_geo_region']
features_df.drop(columns=columns_to_remove, inplace=True)

# Save the filled DataFrame to a new CSV file
features_df.to_csv('filled_features.csv', index=False)

# Now check if there are any null values remaining
print(features_df.isnull().sum())
```

```python
# Load data from combined_data.csv
combined_df = pd.read_csv('combined_data.csv')

# Categorize H1N1 concern levels
# Assuming H1N1 concern levels range from 0 to 3, where 0 represents no concern and 3 represents high concern
# You may adjust the bins according to your actual data
concern_bins = [-1, 0, 1, 2, 3]  # Bins for different concern levels
concern_labels = ['No Concern', 'Low Concern', 'Moderate Concern', 'High Concern']  # Labels for different concern levels
combined_df['concern_level'] = pd.cut(combined_df['h1n1_concern'], bins=concern_bins, labels=concern_labels)

# Grouping the data by concern level and summing the total vaccinated individuals for h1n1 vaccine and seasonal vaccine in each group
vaccinated_by_concern_h1n1 = combined_df.groupby('concern_level')['doctor_recc_h1n1'].sum()
vaccinated_by_concern_seasonal = combined_df.groupby('concern_level')['doctor_recc_seasonal'].sum()

# Plotting the bar plots for h1n1 vaccine and seasonal vaccine by concern level
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Bar plot for h1n1 vaccine by concern level
axs[0].bar(vaccinated_by_concern_h1n1.index, vaccinated_by_concern_h1n1.values, color='lightgreen')
axs[0].set_title('Number of h1n1 Vaccinated Individuals by Concern Level')
axs[0].set_xlabel('Concern Level')
axs[0].set_ylabel('Number of Vaccinated Individuals')
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# Bar plot for seasonal vaccine by concern level
axs[1].bar(vaccinated_by_concern_seasonal.index, vaccinated_by_concern_seasonal.values, color='lightblue')
axs[1].set_title('Number of Seasonal Vaccinated Individuals by Concern Level')
axs[1].set_xlabel('Concern Level')
axs[1].set_ylabel('Number of Vaccinated Individuals')
axs[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
