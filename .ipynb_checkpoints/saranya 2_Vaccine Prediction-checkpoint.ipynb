```python
from sklearn.model_selection import train_test_split
X = v3.drop(["h1n1_vaccine", "seasonal_vaccine"], axis=1)
y_h1n1 = v3["h1n1_vaccine"]
y_seasonal = v3["seasonal_vaccine"]
X_train, X_test, y_train_h1n1, y_test_h1n1, y_train_seasonal, y_test_seasonal = train_test_split(X, y_h1n1, y_seasonal, test_size=0.30, random_state=42)
```

```python

print(X_train.shape)
print(X_test.shape)
print(y_train_h1n1.shape)
print(y_test_h1n1.shape)
print(y_train_seasonal.shape)
print(y_test_seasonal.shape)
```

```python
# Model:LogisticRegression
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
# training model for h1n1
LR.fit(X_train,y_train_h1n1)  
```

```python
# Prediction
y_pred_h1n1 = LR.predict(X_test)
```

```python
# training model for Seasonal vaccine
LR.fit(X_train, y_train_seasonal)
pred_seasonal = LR.predict(X_test)
```

```python
y_pred_seasonal = LR.predict(X_test)
```

```python
# find confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score for models
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
# h1n1_vaccine

acc_h1n1 = accuracy_score(y_test_h1n1, y_pred_h1n1)
prec_h1n1 = precision_score(y_test_h1n1, y_pred_h1n1)
recall_h1n1 = recall_score(y_test_h1n1, y_pred_h1n1)
f1_h1n1 = f1_score(y_test_h1n1, y_pred_h1n1)

# seasonal_vaccine

acc_seasonal = accuracy_score(y_test_seasonal, y_pred_seasonal)
prec_seasonal = precision_score(y_test_seasonal, y_pred_seasonal)
recall_seasonal = recall_score(y_test_seasonal, y_pred_seasonal)
f1_seasonal = f1_score(y_test_seasonal, y_pred_seasonal)
```

```python
print("For h1n1_vaccine:")
print(f"Accuracy: {acc_h1n1}")
print(f"Precision: {prec_h1n1}")
print(f"Recall: {recall_h1n1}")
print(f"F1 Score: {f1_h1n1}")
print("\n")
print("For seasonal_vaccine:")
print(f"Accuracy: {acc_seasonal}")
print(f"Precision: {prec_seasonal}")
print(f"Recall: {recall_seasonal}")
print(f"F1 Score: {f1_seasonal}")
print("\n")
```

```python
CR_h1n1=classification_report(y_test_h1n1, y_pred_h1n1)
print(CR_h1n1)
```

```python
CR_seasonal=classification_report(y_test_seasonal, y_pred_seasonal)
print(CR_seasonal)
```

```python
# 2. Model: RandomForest
from sklearn.ensemble import RandomForestClassifier
RF_CLF = RandomForestClassifier(n_estimators=100)

# Training the model for H1N1

RF_CLF.fit(X_train, y_train_h1n1)

# Predicting

y_pred_h1n1 = RF_CLF.predict(X_test)
```

```python
# Training the seasonal vaccine

RF_CLF.fit(X_train, y_train_seasonal)

# Predicting

y_pred_seasonal = RF_CLF.predict(X_test)
```

```python
# Predictions for seasonal_vaccine
y_pred_seasonal = RF_CLF.predict(X_test)
```
