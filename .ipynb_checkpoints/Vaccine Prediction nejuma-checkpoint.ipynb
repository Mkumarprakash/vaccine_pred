```python
# Importing the Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
pwd
```

```python
#Loading the data
features=pd.read_csv('features.csv')
```

```python
# feature data
features.head()
```

```python
features.tail()
```

```python
# target data
labels = pd.read_csv('labels.csv')
labels.head()
```

```python
# Merging feature and target data
df = features.merge(labels, how='left', on='respondent_id')
df.head()
```

```python
df.columns
```

```python
# checking null values
df.info()
```

```python
df.isna().sum()
```

```python
# columns health_insurance, employment_industry, employment_occupation contain too many NULLs so drop these columns
df.drop(['employment_industry','employment_occupation'],axis=1,inplace = True)
```

```python
df.describe()
```

```python
df_cols = df.select_dtypes(include = 'object').columns
df_cols
```

```python
#All Categorical features which are NULL are substituted with the mode of data, and the Numeric features with NULLs are substituted with the means
```

```python
for col in df.columns:
    if df[col].isnull().sum() and df[col].dtypes != 'object':
        df[col].loc[(features[col].isnull())] = df[col].median()
for col in df.columns:
    if df[col].isnull().sum() and df[col].dtypes == 'object':
        df[col].loc[(df[col].isnull())] = df[col].mode().max()
```

```python
df.isnull().sum()
```

## Model Creation

```python
X=df.drop(['h1n1_vaccine','seasonal_vaccine'], axis=1)
y_h1n1=df['h1n1_vaccine']
y_seasonal = df['seasonal_vaccine']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train_h1n1, y_test_h1n1, y_train_seasonal, y_test_seasonal = train_test_split(X,y_h1n1, y_seasonal, test_size=0.20, random_state=100) 
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train_h1n1)  ## training model for h1n1
```

```python
y_pred_h1n1 = lr.predict(X_test)
```

```python
lr.fit(X_train, y_train_seasonal)## training model for seasonal_vaccine
```

```python
y_pred_seasonal = lr.predict(X_test)
```

## Model Evaluation

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
# h1n1_vaccine

acc_h1n1 = accuracy_score(y_test_h1n1, y_pred_h1n1)
prec_h1n1 = precision_score(y_test_h1n1, y_pred_h1n1)
recall_h1n1 = recall_score(y_test_h1n1, y_pred_h1n1)
f1_h1n1 = f1_score(y_test_h1n1, y_pred_h1n1)
```

```python
# seasonal_vaccine

acc_seasonal = accuracy_score(y_test_seasonal, y_pred_seasonal)
prec_seasonal = precision_score(y_test_seasonal, y_pred_seasonal)
recall_seasonal = recall_score(y_test_seasonal, y_pred_seasonal)
f1_seasonal = f1_score(y_test_seasonal, y_pred_seasonal)
```

```python
print("For h1n1_vaccine:")
print(f"Accuracy: {acc_h1n1}")
print(f"Precision: {prec_h1n1}")
print(f"Recall: {recall_h1n1}")
print(f"F1 Score: {f1_h1n1}")
print("\n")
print("For seasonal_vaccine:")
print(f"Accuracy: {acc_seasonal}")
print(f"Precision: {prec_seasonal}")
print(f"Recall: {recall_seasonal}")
print(f"F1 Score: {f1_seasonal}")
print("\n")
```

```python
cr_h1n1=classification_report(y_test_h1n1, y_pred_h1n1)
print(cr_h1n1)
```

```python
cr_seasonal=classification_report(y_test_seasonal, y_pred_seasonal)
print(cr_seasonal)
```
